Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```



*   Uploading Dataset
*   Load the dataset using pandas.




```python
df=pd.read_csv('/content/IRIS.csv')
```

  

 ●	Print the shape, column names, and the first few rows using .head().




```python
print('Shape of Dataset',df.shape)
print('\n Column Names',df.columns)
print('\n First Few Rows',df.head(10))
```

●	Use .info() and .describe() for summary statistics.

```python
print('Information of Dataset')
df.info()
print('\n Summary Statistics')
df.describe()
```

●	Visualize the dataset:

○ Create a scatter plot to show relationships between features.



```python
plt.scatter(df['sepal_length'], df['petal_length'], color='purple', label='Sepal Length vs Petal Length')
plt.xlabel('sepal_length')
plt.ylabel('petal_length')
plt.title('Scatter Plot of Sepal Length vs Petal Length')
plt.legend()
plt.show()
```

○ Use histograms to show value distributions.

```python
plt.hist(df['sepal_length'], color='pink', edgecolor='black')
plt.xlabel('sepal_length')
plt.ylabel('Frequency')
plt.title('Histogram of Sepal Length')
plt.show()
```

○ Use box plots to identify outliers.

```python
plt.boxplot([df['sepal_length'], df['petal_length'], df['petal_width'], df['petal_width']])
plt.ylabel('Values')
plt.title('Box Plot of IRIS Dataset')
plt.xticks([1, 2, 3, 4], ['Sepal Length', 'Petal Length','Petal Length', 'Sepal Length'])
plt.show()
```
